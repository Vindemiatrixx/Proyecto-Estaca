## *features*
-  **Neo Reference ID**: Identificador único para cada objeto cercano a la Tierra (NEO).

-  **Name**: Identificador adicional para el objeto cercano a la Tierra.

-  **Absolute Magnitude**: Medida de la luminosidad intrínseca de un objeto celeste.

-  **Est Dia in KM(min)**: Diámetro mínimo estimado del asteroide en kilómetros.

-  **Est Dia in KM(max)**: Diámetro máximo estimado del asteroide en kilómetros.

-  **Est Dia in M(min)**: Diámetro mínimo estimado del asteroide en metros.

-  **Est Dia in M(max)**: Diámetro máximo estimado del asteroide en metros.

-  **Est Dia in Miles(min)**: Diámetro mínimo estimado del asteroide en millas.

-  **Est Dia in Miles(max)**: Diámetro máximo estimado del asteroide en millas.

-  **Est Dia in Feet(min)**: Diámetro mínimo estimado del asteroide en pies.

-  **Est Dia in Feet(max)**: Diámetro máximo estimado del asteroide en pies.

-  **Close Approach Date**: Fecha en la que el asteroide estará más cerca de la Tierra durante su aproximación.

-  **Epoch Date Close Approach**: Fecha de la aproximación del asteroide en formato de época.

-  **Relative Velocity km per sec**: Velocidad relativa del asteroide con respecto a la Tierra en kilómetros por segundo.

-  **Relative Velocity km per hr**: Velocidad relativa del asteroide con respecto a la Tierra en kilómetros por hora.

-  **Miles per hour**: Velocidad relativa del asteroide con respecto a la Tierra en millas por hora.

-  **Miss Dist.(Astronomical)**: Distancia mínima a la que el asteroide pasará de la Tierra en unidades astronómicas.

-  **Miss Dist.(lunar)**: Distancia mínima a la que el asteroide pasará de la Tierra en distancias lunares.

-  **Miss Dist.(kilometers)**: Distancia mínima a la que el asteroide pasará de la Tierra en kilómetros.

-  **Miss Dist.(miles)**: Distancia mínima a la que el asteroide pasará de la Tierra en millas.
